### What kind of information can you not infer from the link list representation of the network that you can infer from the adjacency matrix?

**The number of nodes in the graph.**   
If we know the adjacency matrix, we can infer how many nodes the graph have by looking at the length of the matrix. However, we can't infer from the link list. For example, we cannot infer that there is a node5 in image2.20a from the link list.

### In the (a) network, how many paths (with possible repetition of nodes and links) of length 3 exist starting from node 1 and ending at node 3? And in (b)?

```python
# all_simple_paths(G, source, target, cutoff=None)
# cutoff:Depth to stop the search. Only paths of length <= cutoff are returned.
count = 0
for path in nx.all_simple_paths(UG, 1, 3, cutoff=3):
    if len(path) == 4:
        count += 1
        print(path)
if count == 0:
    print("no such path")
```

## 2.5 Bipartite Networks

Consider the bipartite network of Image below


### Construct its adjacency matrix. Why is it a block-diagonal matrix?

```python
from networkx.algorithms import bipartite
```

```python
B = nx.Graph()
edges_B = [(1,7),(2,9),(3,7),(3,8),(3,9),(4,9),(4,10),(5,9),(5,11),(6,11)]
b1 = list(range(1,7))
b2 = list(range(7,12))
# Add nodes with the node attribute "bipartite"
B.add_nodes_from(b1, bipartite=0)
B.add_nodes_from(b2, bipartite=1)
# Add edges only between nodes of opposite node sets
B.add_edges_from(edges_B)

# set positions for bipartite nodes
pos = dict()
pos.update( (n, (1, i)) for i, n in enumerate(b1) )
pos.update( (n, (2, i+0.5)) for i, n in enumerate(b2) )

color_list = ['purple','purple','purple','purple','purple','purple','Green','Green','Green','Green','Green']
nx.draw(B,with_labels=True,pos=pos,font_weight='bold',font_color='w',node_color = color_list)
```

### Construct the adjacency matrix of its two projections, on the purple and on the green nodes, respectively.

```python
R = nx.Graph()
b1_pro = [(1,3),(2,3),(2,4),(2,5),(3,4),(3,5),(4,5),(5,2),(5,6)]
R.add_nodes_from(b1)
R.add_edges_from(b1_pro)
nx.adjacency_matrix(R).todense()
```

```python
R2 = nx.Graph()
b2_pro = [(7,8),(7,9),(8,9),(9,10),(9,11)]
R2.add_nodes_from(b2)
R2.add_edges_from(b2_pro)
nx.adjacency_matrix(R2).todense()
```

### Calculate the average degree of the purple nodes and the average degree of the green nodes in the bipartite network.

```python
d1 = bipartite.degrees(B,b1)[0]
d2 = bipartite.degrees(B,b1)[1]
avg_d1 = np.mean([v for k,v in d1])
avg_d2 = np.mean([v for k,v in d2])
print(d1,"'s average degree is",avg_d1)
print(d2,"'s average degree is",avg_d2)
```

### Calculate the average degree in each of the two network projections. Is it surprising that the values are different from those obtained in point (c)?

```python
print(np.mean([v for v in R.degree()]))
print(np.mean([v for v in R2.degree()]))
```

### Binning integers

**Generate 5000 numbers drawn from a Poisson distribution characterized by Î»=10**    
**Find the maximum and minimum value**
- Poisson distribution:
    - a discrete probability distribution
    - a fixed interval of time or space
    - occur with a known constant mean rate and independently
    
- Examples
    - The Poisson distribution may be useful to model events such as
    - The number of meteorites greater than 1 meter diameter that strike Earth in a year
    - The number of patients arriving in an emergency room between 10 and 11 pm
    - The number of laser photons hitting a detector in a particular time interval

```python
psdis = np.random.poisson(10, 5000)
psmin, psmax = np.min(psdis), np.max(psdis)
print("Poisson distribution:", psdis, "\n" "minimum:", psmin, "\n" "maximum:", psmax)
```

**Specify the bins& the sum over bin counts**

```python
v = np.arange(psmax + 1)
pshist = np.histogram(psdis, bins=v)
print(pshist)
```

**Use a bar chart (matplotlib.pyplot.bar) to plot the distribution**

```python
x = np.arange(psmin, psmax)
hist = np.histogram(psdis, bins=x)
plt.bar(hist[1][:-1], hist[0], width=0.8)
```

### Binning and plotting degree distributions.

**ER parameters (N and p)**   


```python
# p=â¨kâ©/(N-1)
N = 5000
k = 10
p = k / (N - 1)
```

**Use networkx to create the graph and extract the degree distribution.**

```python
er = nx.erdos_renyi_graph(N,p)
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.title('the degree distribution of ER network')
    plt.xlabel('degree')                                                                                                             
    plt.ylabel('frequency')
    plt.show()

plot_degree_dist(er)
```
