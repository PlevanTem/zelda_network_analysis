### What's the problem with random networks as a model for real-world networks according to the argument in section 3.5 (near the end)?
As it turns out, most large networks in the present day have scale-free structure, or a network structure in which most nodes have few connections, some nodes have a medium number, and a few nodes are very well connected. Using Twitter as an example, most of us may have a few hundred followers, some of us – Andrew Aquino, for example – have a few thousand, and Justin Bieber has millions.

However, the significant deviation between the data and the Poisson fit indicates that the random network model underestimates the size and the frequency of the high degree nodes, as well as the number of low degree nodes. In conclusion, many nodes in random network have similar degree distribution without hubs, while there are many highly-connected nodes in real network and there are huge distinction betweem their degrees.

```python
import networkx as nx
from matplotlib import pyplot as plt
import numpy as np
```

```python
# generate 3 graphs with 500 nodes each, average degree = 4
# rewiring probablity p=0,0.1, and 1. 
G1 = nx.watts_strogatz_graph(500,4,p=0)
G2 = nx.watts_strogatz_graph(500,4,p=0.1)
G3 = nx.watts_strogatz_graph(500,4,p=1)
```

```python
# Calculate the average shortest path length ⟨d⟩ for each one. 
d1 = nx.average_shortest_path_length(G1)
d2 = nx.average_shortest_path_length(G2)
d3 = nx.average_shortest_path_length(G3)
print('the average shortest path length ⟨d⟩ for G1:',d1)
print('the average shortest path length ⟨d⟩ for G2:',d2)
print('the average shortest path length ⟨d⟩ for G3:',d3)
```

As we can see, the average shortest path for these 3 graphs shortens with the increasing of p. When p = 1, the <d> for G1(about 62.9) sharply decresed to about 4.9

### Generate 50 networks with N=500, ⟨k⟩=4, for each of p={0,0.01,0.03,0.05,0.1,0.2}.

```python
p = np.array([0,0.01,0.03,0.05,0.1,0.2])
names = locals()
d_set = np.array([])
for i in range(6):
    names['G'+str(i)] = nx.watts_strogatz_graph(500,4,p=p[i%6])
    names['d'+str(i)] = nx.average_shortest_path_length(names['G'+str(i)])
    d_set = np.append(d_set,[names['d'+str(i)]])
    
plt.plot(p,d_set,'o',linestyle='solid',color='black')
plt.ylim(0,70)
plt.xlabel('p')
plt.ylabel('ave(<d>)')
std = np.std(d_set)
plt.errorbar((0.01),(std),yerr=6,color='black',capsize = 3)
plt.show()
```

### What are the three slope dependent regimes of complex networks with power-law degree distributions? Briefly describe each one. (You will have to skim chp 4.7 to answer this one).

1. Anomalous Regime (γ≤ 2)   
For γ ‹ 2 the exponent 1/(γ− 1) in (4.18) is larger than one, hence the number of links connected to the largest hub grows faster than the size of the network, which shouldn't happen.

2. Sale-free Regime (2 ‹ γ ‹ 3)   
 kmax grows with the size of the network with exponent 1/(γ - 1), which is smaller than one, while the market share of the largest hub, kmax /N, representing the fraction of nodes that connect to it, decreases as kmax/N ~ N-(γ-2)/(γ-1)

3. Random   
Scale-free networks with large γ are hard to distinguish from a random network.


### What are the three regimes we find in non-linear preferential attachement? (chapter 5) Briefly describe each one.

1. Sublinear Regime (0 < α < 1)   
The degree distribution follows the stretched exponential, resulting in fewer and smaller hubs than in a scale-free network. As α → 1 the cutoff length increases and pk follows a power law over an increasing range of degrees.

2. Linear Regime (α=1)   
This corresponds to the Barabási-Albert model, hence the degree distribution follows a power law.

3. Superlinear Regime (α > 1)   
The high-degree nodes are disproportionately attractive. A winner-takes-all dynamics leads to a hub-and-spoke topology. In this configuration the earliest nodes become super hubs and all subsequent nodes link to them.

### Create our own Barabasi-Albert model (a special case)

```python
import networkx as nx
import random
```

```python
G = nx.Graph()
G.add_nodes_from([1,2,3])
G.add_edges_from([(1,2),(2,3)])
nx.draw(G, with_labels=True, font_weight='bold',font_color='white')
```

```python
# create a 100-node Barabasi_Albert network
def generate_Barabasi_Albert(G,node):
    flatten = np.array([])
    #flatten the list of edges
    for i in G.edges:
        for j in i:
            flatten = np.append(flatten,j)
    G.add_node(node)
    G.add_edges_from([(node,random.choice(flatten)),(node,random.choice(flatten))])
for i in range(4,101):
    generate_Barabasi_Albert(G,i)
```

```python
nx.draw(G,node_size=300,with_labels=True, font_size=10, font_color='white',width=0.4)
```

```python
# Add more nodes until you have a 5000 node network.
for i in range(101,5001):
    generate_Barabasi_Albert(G,i)
```

```python
degrees = np.array([G.degree(n) for n in G.nodes()])
max_degree = np.max(degrees)
min_degree = np.min(degrees)
```

### explore the Friendship paradox. 

```python
times_of_true = 0
for i in range(1000):
    i = random.choice(list(G.nodes()))
    # the number of i's friends
    i_degree = G.degree(i)
    # find i's neighbors
    nb = list(G.neighbors(i))
    # calculate the avg degree of i's friends
    degree_of_friends = np.array([])
    for i in nb:
        degree_of_friends = np.append(degree_of_friends,G.degree(i))
    avg_degree_of_friends = np.mean(degree_of_friends)
    if i_degree < avg_degree_of_friends:
        times_of_true+=1
```
