# Part 1:  Binning degree distributions
##  Binning real numbers
**1.1)**  Let's do a gentle start and use the random library to generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
from IPython import display
import scipy as sp
from numpy import math
from random import choice
mu, sigma = 2, 0.125 # mean and standard deviation
RanGauss = np.random.normal(mu, sigma, 5000)
```

**1.2)** Now, let's use numpy.histogram to bin those number into 10 bins. What does the numpy.histogram function return? Do the two arrays have the same length?

    Answer to 1.2): The numpy histogram function takes a list of observations as the first input, and an integer or a list of numbers as the second input. It then counts how many of the observations in the first input there exists within the interval between two entries in the second input and then makes a barplot. If the second input is an integer k, the function will allocate an equal distance of k intervals on the x axis.  The barplot will show how many elements from input 1, has the numerical value that falls within a given of the k intervals. 
    
    In this particular case,the random gaussian array has length 5000, because there are 5000 observations, whereas using 10 bins means using an array of length 10, so the two arrays do not have the same length.

```python
a = min(RanGauss)
b = max(RanGauss)
CustomBins = list(np.arange(a,b+((b-a)/10),((b-a)/10)))
fig = plt.figure()
hist1 = fig.add_subplot(2, 1, 1)
hist2 = fig.add_subplot(2, 1, 2)
hist1.text(a,1000,'Bins = 10')
hist1.hist(RanGauss,10)
hist2.text(a,1000,'Bins = CustomBins')
hist2.hist(RanGauss,CustomBins)
plt.show()
```

**1.3)** When we use matplotlib.pyplot.plot to plot the binned data. You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.

    Answer to 1.3): One can deal with this issue by creating a bin list of any desired length,depending on how granular the intervals one wants to group the observations in. With use of the range() for integers or the numpy.arange() for floating numbers, one can create bins, starting with the minimum observed value and ending with the maximum value, and then divide the distance between the two with a discrete amount depending on how many bins one wishes to create. In the example 'CustomBins' above a bin list of length 10 was created by dividing the distance by 10.

##  Binning integers

**1.4)** But binning real numbers into a fixed number of bins is easy when numpy.histogram does all the work and finds the right bin boundaries for you. Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out numpy also has some convenient random number generators. Use numpy.random.poisson to generate 5000 numbers drawn from a Poisson distribution characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers. 

    Answer to 1.4): see the code below

```python
pois1 = np.random.poisson(10,5000)
a = min(pois1)
b= max(pois1)
print("min = ", a,"max = ", b)
```

**1.5)**
Instead of simplify specifying the number of bins for numpy.histogram, let's specify the bins we want using a vector. Create a vector $v$ that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, and the last bin contains the maximum number (you may want to think about how numpy.histogram deals with the smallest an largest value in particular;

a) is it correct to set the first entry of $v$ to be equal to the smallest value you observe and the last entry of $v$ to be the highest value observed?). Use the vector by setting numpy.histogram's bin parameter as bin = $v$.

b) What is the sum over bin counts?

c) Explain how the binning-vectors first and last element relates to the min and max from the Poisson distribution.

    Answer to 1.5): see the code below
      1.5.a) it is not correct to set the last entry of the vector v to be the highest value observed, instead add an extra entry after the hightest value observed and define it as "max + (max-min)/max" (when dealing with integer observations).  
     1.5.b) As can be seen below, the approach described in the answer to 1.5.a) leads to the correct sum over the bin counts.
     1.5.c) The binning-vectors first entry should be the minimum value observed, the second last entry of the binning-vector should be the highest value observed so these values are strongly connected to the min and max from the Poisson distribution.

```python
v = list(np.arange(a,b,((b-a)/b)))#the bin-vector with the highest value observed as the last entry
v2= list(np.arange(a,b+((b-a)/10),((b-a)/b))) #the bin-vector with the highest value observed as the second last entry
HiPoi = plt.hist(pois1, v)#histrograms are defined so we can do the sum in the next chunk of code
HiPoi2 = plt.hist(pois1, v2)
```

```python
number_of_bins1 = len(v)
binWidth = len(HiPoi[1][0:1])
a_sum = sum(HiPoi[0][0:number_of_bins1]*binWidth)
number_of_bins2 = len(v2)
binWidth2 = len(HiPoi2[1][0:1])
b_sum = sum(HiPoi2[0][0:number_of_bins2]*binWidth2)
print("answer to 1.5.b), if the last endpoint in v is the highest value observed we get the sum over bin counts to be ", a_sum," if the last entry of v is max +(max-min)/max we get", b_sum)
```

**1.6)** Now, use a bar chart (matplotlib.pyplot.bar) to plot the distribution 

```python
y = []
X0 = []
iterList = list(range(min(pois1), max(pois1)+1,1))
for x in iterList:
    y = y + [list(pois1).count(x)]
    X0 = X0 + [x]

plt.title("Last entry of v = highest value observed", fontdict = {'fontsize':10})
plt.suptitle(" Bar plot of the random poisson distribution", fontdict = {'fontsize':15})
plt.xlabel('Degree of vertices in G')
plt.ylabel('Amount of vertices in G')
plt.bar(X0,y)
plt.show()
```

## Binning and plotting degree distributions. 

**1.7)** Let's generate the Erd√∂s-Renyi (ER) network which has a degree distribution that matches the Poisson distribution above. First we have to figure out which values the ER parameters (N and p) should assume. It's easy to see that $N = 5000$, but how do you find $p$? Hint: The parameter $\lambda$ in the Poisson distribution corresponds to the average degree, so you have to find a $p$ that results in an average degree, $k = 10$. And you know that $\langle k \rangle = p (N-1)$, which will give you $p$.

   Answer to 1.7) we see that $p(N-1)=p(4999)=10 \implies p= \frac{10}{4999}$ so our random graph needs to generate new edges with probability $ p=\frac{10}{4999}$ this should give us an average degree of 10 edges per node in the random graph of 5000 vertices. 

**1.8)** Now, use networkx to create the graph and extract the degree distribution.

    Answer to 1.8) In the below code we generate a random graph with the above defined p value and see an average degree of approximately 10

```python
from numpy import math
G= nx.erdos_renyi_graph(5000,10/4999)#use p(N-1)=10 iff p=10/4999
g_deg = nx.degree_histogram(G)
g_deg_sum = [a*b for a,b in zip(g_deg, range(0,len(g_deg)))]
print('average degree: {}'.format(sum(g_deg_sum) / G.number_of_nodes()))
```

**1.9)** Finally, create a nice bar plot of the degree distribution, including axes labels and a plot title. Make sure that it looks like the Poisson distribution you plotted above.

```python
iterList2 = list(range(min(g_deg),max(g_deg),1))
degrees = [val for (node, val) in G.degree()]
y = []
X1 = []
for x in iterList2:
    if degrees.count(x) > 0:
        y = y + [degrees.count(x)]
        X1 = X1 + [x]
plt.title(r'G(N,p)=G(5000,$\frac{10}{4999}$) degree distribution of vertices')
plt.xlabel('Degree of vertices in G')
plt.ylabel('Amount of vertices in G')
plt.bar(X1,y)
plt.show()
```

# Part 3: Watts-Strogatz Networks

**3.1)** Use nx.watts_strogatz_graph to generate 3 graphs with 500 nodes each, average degree = 4, and rewiring probablity $p = 0, 0.1, \textrm{and} 1$. Calculate the average shortest path length $\langle d \rangle$ for each one. Describe what happens to the network when $p = 1$.

    Answer to 3.1) as the simulation below reveals, the closer the probability p gets to 1 the closer the average path lenght in the watts_strogatz graph will approach the average degree of each node, in this example the average path length will approach the value 4.

```python
p_list = [0,0.1,1]
G1 = nx.watts_strogatz_graph(500,4,0)
d_1 = nx.average_shortest_path_length(G1)
G2 = nx.watts_strogatz_graph(500,4,0.1)
d_2 = nx.average_shortest_path_length(G2)
G3 = nx.watts_strogatz_graph(500,4,1)
d_3 = nx.average_shortest_path_length(G3)
list1 = [d_1,d_2,d_3]
for i in range(3):
    print('avg length of shortest path for network with avg degree 4, and probability p =',p_list[i],' is  = ', list1[i])
```

**3.2** Let's understand the behavior of the WS model as we increase p in more detail. Generate 50 networks with $N = 500$, $\langle k \rangle = 4$, for each of $p = \{0, 0.01, 0.03, 0.05, 0.1, 0.2\}$. Calculate the average of $\langle d \rangle$ as well as the standard deviation over the 50 networks, to create a plot that shows how the path length decreases very quickly with only a little fraction of re-wiring. Use the standard deviation to add errorbars to the plot. My version of the plot is below (since a picture's worth 1000 words).

    Answer 3.2) The simulation of the graphs is run in the first code chunk. In the next chunk the plot is created, however the standard deviation was so small that the list was printed in the third chunk as a sanity check, since the standard deviation was hardly visible for most p-values on the plot.

```python
p_0List = []
p_1List = []
p_2List = []
p_3List = []
p_4List = []
p_5List = []
for i in range(50):
    G1 = nx.watts_strogatz_graph(500,4,0)
    d_1 = nx.average_shortest_path_length(G1)#np.log(500)/np.log(k_1)
    p_0List.append(d_1)
    G2 = nx.watts_strogatz_graph(500,4,0.01)
    d_2 = nx.average_shortest_path_length(G2)#np.log(500)/np.log(k_2)
    p_1List.append(d_2)
    G3 = nx.watts_strogatz_graph(500,4,0.03)
    d_3 = nx.average_shortest_path_length(G3)#np.log(500)/np.log(k_3)
    p_2List.append(d_3)
    G4 = nx.watts_strogatz_graph(500,4,0.05)
    d_4 = nx.average_shortest_path_length(G4)#np.log(500)/np.log(k_4)
    p_3List.append(d_4)
    G5 = nx.watts_strogatz_graph(500,4,0.1)
    d_5 = nx.average_shortest_path_length(G5)#np.log(500)/np.log(k_5)
    p_4List.append(d_5)
    G6 = nx.watts_strogatz_graph(500,4,0.2)
    d_6 = nx.average_shortest_path_length(G6) #np.log(500)/np.log(k_6)
    p_5List.append(d_6)
superList = [p_0List,p_1List, p_2List, p_3List, p_4List, p_5List]
avg_d_list = []
st_dev = []
p_list = [0,0.01,0.03,0.05,0.1,0.2]
for x in superList:
    a1 = 1/50*sum(x)#the mean of the path lenght
    avg_d_list.append(a1)#append mean of path length
    burner_list = []
    for y in x:
        a = (y-a1)*(y-a1)
        burner_list.append(a)#append the variance in the mean of the path lenght
    st_dev.append(np.sqrt((1/50)*sum(burner_list)))#append the standard deviation
```

```python
plt.figure(3,figsize=(10,6))
plt.title('Watts strogatz distribution  50 simulations of 5 different graphs G(n,k,p), with n= 500, k=4 and p varies on x axis')
plt.xlabel('p')
plt.ylabel('avg path length <d>')   
#plt.plot(p_list,avg_d_list,'bo')  
plt.plot(p_list,avg_d_list) 
for i in range(len(st_dev)):#insert standard deviation for each p
    yerr = np.linspace(st_dev[i], st_dev[i], 1)
    plt.errorbar(p_list[i], avg_d_list[i],yerr=yerr, label= None)
```

# Part 5 Power-laws and the friendship paradox

Next step is to explore the [Friendship paradox](https://en.wikipedia.org/wiki/Friendship_paradox). This paradox states that almost everyone have fewer friends than their friends have, on average. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution. The explanation is that almost everyone is friends with a hub, that drives up the average degree of the friends. Let's explore that in the 5000 node BA network we've just generated. Do the following:

**5.1)** Pick a node i at random (e.g. use random.choice). Find its degree.

```python
from random import choices
list4 = list(G1.nodes())
random2= [choices(list4,k=1)]
deg_1 = G1.degree(random2[0][0])
print("node ", random2[0][0], " has degree ", deg_1)
```

**5.2)** Find $i$'s neighbors. And calculate their average degree.

```python
NB = list(G1.neighbors(random2[0][0]))
deg_2 = 1/(len(degrees))*sum(degrees)
print("node ", random2[0][0], " with degree ", deg_1, " has neighbours ", NB, " with avg. degree ", deg_2)
```

 **5.3)** Compare the two numbers to check if it's true that i's friends (on average) have more friends than $i$.

```python
print("In this particular random example it is", deg_1< deg_2, "that the node ", random2[0][0], " has a lower degree (fewer friends) than the average of its neighbours" )
```

**5.4)** Do this 1000 times. How many out of those 1000 times is the friendship paradox true?

```python
from random import choices
list4 = list(G1.nodes())
resultList = []
for i in range(1000):
    random2= [choices(list4,k=1)]
    NB = list(G1.neighbors(random2[0][0]))
    deg_1 = G1.degree(random2[0][0])
    degrees = [val for (node, val) in G1.degree(NB)]
    deg_2 = 1/(len(degrees))*sum(degrees)
    if deg_2 > deg_1:
        resultList.append(1)
    else:
        resultList.append(0)
print('The friendship paradox was true for ', 100*sum(resultList)/1000, '% of the cases ie.', sum(resultList), " of 1000 cases")
```
