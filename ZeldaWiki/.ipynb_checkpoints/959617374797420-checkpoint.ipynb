### Binning real numbers

```python
import networkx as nx
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Creating 5000 random datapoints with gaussian distribution 
mu = 2
sigma = 0.125
s = [random.gauss(mu, sigma) for i in range(5000)]
```

```python
# Binning the numbers 
array, bins = np.histogram(s, bins = 10)
print("length of array = " ,len(array))
print("length of bins = " ,len(bins))
```

The arrays are not the same length. Bins ends at the last value, which means that when we plot it we need to only use the first 10 values.   
if we are not just using the 10 first values, then it wouldn't be able to plot

```python
# Plotting the data 
plt.plot(bins[:-1], array)
plt.xlabel("Bins")
plt.ylabel("count")
```

Th eplot show the gaussian distribution with $\mu$ =2 and $\sigma$ = 0.125

### Binning integers

```python
# Creating an array with 5000 random values from a poisson distribution with lambda = 10
lam = 10
s= np.random.poisson(lam, 5000)
print("max =", max(s), ", min = ", min(s))
```

Creating an array with custom bins.  
I want 10 bins

```python
v = np.arange(min(s), max(s)+min(s), (max(s)-min(s))/10)
print(v)
```

As it can be seen the vector has length 11. This is because we have the last value. This is not a value that we want when we are plotting, since we don't have any values in our array greater than 23.  
Numpy's bin vector would look exactly the same way, if you specify 10 bins. 

```python
hist, _ = np.histogram(s, bins=v)
hist, v
```

```python
print("sum of bins count = ", sum(hist))
```

Plotting the bins

```python
plt.bar(v[:-1], hist)
plt.xlabel("Bins")
plt.ylabel("count")
```

### Binning and plotting degree distributions

Generating an ER newtork

We have to find p. It is known that $<k> = p(N-1)$  
This means that p can be isolated and $p = \dfrac{k}{N-1}$

```python
import networkx as nx
```

```python
N = 5000
k = 10
p = k/(N-1)
print("p =" , p)
```

```python
# Generating the network 
G = nx.erdos_renyi_graph(N, p)
```

```python
# Getting the degree distribution 
degrees = [G.degree(n) for n in G.nodes()]

# Plotting the historgram    
plt.hist(degrees)
plt.title("Degree distribution of erdos-renyi network (N=5000, k=10)")
plt.xlabel("Degree distribution")
plt.ylabel("count")
plt.show()
```

# Part 3

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
k = 4
p_list = [0, 0.1, 1]

for p in p_list:
  ws = nx.watts_strogatz_graph(N, k, p)
  print(f'p = {p}, <d> = {nx.average_shortest_path_length(ws)}')
```

With the Watts-Strogatz model we create a ring-shape network with n nodes where each node is connected to k nearest neighbour (k-1 neighbors if k is odd). By increasing probability p we go from a regular lattice (p=0), which has high clustering but lacks the small-world phenomenon to a random network (p=1).

```python
N = 500
k = 4
p_list=[0,0.01,0.03,0.05,0.1,0.2]
avg_d = []
std_d = []
for p in p_list:
  tmp = []
  for i in np.arange(0,50):
    ws = nx.watts_strogatz_graph(N, k, p)
    tmp.append(nx.average_shortest_path_length(ws))
  avg_d.append(np.mean(tmp))
  std_d.append(np.std(tmp))
```

```python
fig = plt.figure(figsize=(12,8))
plt.errorbar(p_list, avg_d, yerr=std_d, marker='o')
plt.xlabel('p',fontsize=10)
plt.ylabel('avg(<d>)',fontsize=10)
```

```python
import networkx as nx
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from([1,2])
G.add_edges_from([(1,2)])
#creating a network with 100 nodes
for i in range(1,100,1):
    G.add_node(i)
    edges=np.array(list(G.edges)).flatten()
    edge=random.choice(edges)
    G.add_edge(edge,i)
nx.draw(G,node_size=70)
```

```python
G = nx.Graph()
G.add_nodes_from([1,2])
G.add_edges_from([(1,2)])
#Creating a network with 5000 nodes
for i in range(1,5000,1):
    G.add_node(i)
    edges=np.array(list(G.edges)).flatten()
    edge=random.choice(edges)
    G.add_edge(edge,i)

```

```python
#Finding the degree of every node in G
degree_sequence = [d for n, d in G.degree()]
dmax = max(degree_sequence)
dmin = min(degree_sequence)
print("The maximum degree of a node in G is: "+str(dmax))
print("The minimum degree of a node in G is: "+str(dmin))
```

```python
hist,bins =np.histogram(degree_sequence, bins=500)
fig, axs = plt.subplots(2)
axs[0].plot(bins[:-1],hist,'o',markersize=2)
axs[1].plot(bins[:-1],hist,'o',markersize=2)
plt.yscale('log')
plt.xscale('log')
plt.xlabel("k")
plt.ylabel("Count")
plt.show()
```

# Part 5

```python
#Picking a random node from G
node=random.choice(list(G.nodes()))
print("The degree of the randomly chosen node is: " + str(G.degree[node]))

#finding the degrees of the neighbouring nodes and calculating the average
degree=([G.degree[n] for n in G.neighbors(node)])
avg=np.mean(degree)
print("The average number of friends for the neighbors of the chosen node is: "+str(avg))

#Doing the above a thousand times
count =0
for i in range(1000):
    node=random.choice(list(G.nodes()))
    degrees =[G.degree[n] for n in G.neighbors(node)]
    avg=np.mean(degrees)
    if (G.degree(node)<avg):
        count +=1
print("The friendship paradox is true "+str(count)+" times ie. " + str(count/1000)+"% of the time")
```

```python
#Creating an erdos renyi network with 5000 nodes 
N=5000
p=N/(N*(N-1)/2)
ER= nx.erdos_renyi_graph(N,p)
```

```python
def plot_degree_distribution(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()
plot_degree_distribution(G)
plt.figure(2)
plot_degree_distribution(ER)
plt.show()

print("The number of connected components in the Barabasi-Albert network is: "+ str(nx.number_connected_components(G)))
print("The number of connected components in the Erdos Renyi network is: "+ str(nx.number_connected_components(ER)))
```

The difference between the networks can be seen in the above histograms. The number of nodes conneted to few other nodes are significantly larger in the Barabasi-Albert network. Furthermore All of the components are connected in the Barabasi-Albert network, which is not the case in the Erdos-Renyi network. The connceted components could also be seen as sub networks, which means there are 877 different node connections. This is best seen by plotting the network, but since a network of size 5000 is visually hard to intepret, a network with 50 nodes is plotted below. From the figure it can easily be seen that the network has more than 1 connceted component (there is not a link between every node).
