To create our network, we downloaded the Zelda Wiki pages for all characters in BotW (during Week 4) and linked them via the hyperlinks connecting pages to each other. To achieve this goal we have used regular expressions!

> * Explain the strategy you have used to extract the hyperlinks from the Wiki-pages, assuming that you have already collected the pages with the Zelda API.

> <font color='#074cad'> After collecting the pages with the Zelda API, the hyperlink extraction was conducted by looping through the Zelda characters. For each character, the corresponding Wiki text was loaded as a string. Next the regular expressions tool was used in order to identify the hyperlinks on different formats, e.g. {{Term|BotW|name of character|link}} and [[name of character]], and subsequently extract the name of the linked character. </font>

> * Show the regular expression(s) you have built and explain in details how it works.

> <font color='#074cad'> The following RegEx patterns were used in order to detect character names only from links in different formats in each file:<br>
    `\{\{Term\|BotW\|(.*?)(?:\|.*?)?\}\}`<br> 
    `\[\[(.*?)(?:\|.*?)?]\]`<br><br>
Whenever either `. ^ $ * + ? { [ ] \ | ( )` is included as a symbol to identify in a text, a `\` must predeed it in the search pattern, because they all have special functions in RegEx. Hence the above patterns will match expressions like {{Term|BotW| *character link*}} and [[character link]].<br>
>To extract names from text in the search pattern `(.*?)(?:\|.*?)?` each function must be understood.<br>
Following definitions from https://developers.google.com/edu/python/regular-expressions:<br>
> * `.` matches any single character except newline '\n'
> * `*` match 0 or more occurrences of the pattern to its left
> * `?` match 0 or 1 occurrences of the pattern to its left. If you add a ? at the end of a pattern, such as .*?, the match will be non-greedy, i.e. stop as soon as it can and not trying to fit the largest possible content.
> * `( )` is a "group" feature of a regular expression. The parenthesis do not change what the pattern will match, instead they establish logical "groups" inside of the match text.</font>

><font color='#074cad'> The `(.*?)`-pattern will thus match all letters/symbols untill it's stopped by either a string matching `(?:\|.*?)?` or the }} / ]]
The `(?:\|.*?)`-pattern will match  "|" and all succeeding letters/symbols. However, the initial `?:`-part constitute an "obscure optional feature", meaning this is a group in the pattern which should be matched but not extracted.<br>
Adding the final `?` to above pattern will then match if the final "|"-part is present as well as if it is not.<br><br>
All of the patterns must be matched, but not all of the string for the RegEx to extract the matches. These search patterns has thereby extracted character *names* only in e.g. {{Term|BotW|name of character|link}} and [[name of character]].
</font>



# Part 1: Network visualization and basic stats

Visualize the network (from lecture 5) and calculate stats (from lecture 4 and 5). For this exercise, we assume that you've already generated the BotW network and extracted the giant connected component. Use the GCC to report the results.

> <font color='#074cad'> To proceed with the stats and visualization we first generate the BotW network and extract the giant connected component (GCC).  </font>

```python
# read csv file provided by Anna
df = pd.read_csv("ZeldaBotW_nodes_w_attributes_enriched.csv")
```

```python
df.head()
```

```python
# replace white space with underscore in all character names
df["Name"] = df["Name"].replace(" ","_",regex=True)
df["Name"] = df["Name"].replace(r"'","_",regex=True)
```

```python
# creating directed graph and adding character names as nodes
DG = nx.DiGraph()
DG.add_nodes_from(df.Name)

node_attr = df.set_index('Name').to_dict('index')
nx.set_node_attributes(DG, node_attr)
```

```python
# adding edges to the network based on character files                
pattern1 = "\[\[(.*?)(?:\|.*?)?]\]"
pattern2 = "\{\{Term\|BotW\|(.*?)(?:\|.*?)?\}\}"
pattern3 = "\w+"

for i in df.Name:
    with open(f"{char_path}{i}.txt","r") as f:
        node_links = f.read()
        words = re.findall(pattern3, node_links)
        link_list1 = re.findall(pattern1, node_links)
        link_list2 = re.findall(pattern2, node_links)
        DG.nodes[i]["word_len"]=len(words)
        for j in link_list1:
            j = j.replace(' ','_')
            if j in df.Name.values:
                DG.add_edge(i,j)
        for k in link_list2:
            k = k.replace(' ','_')
            if k in df.Name.values:
                DG.add_edge(i,k)
```

```python
# extracting the GCC
largest_cc = max(nx.weakly_connected_components(DG), key=len)
sub_G = DG.subgraph(largest_cc).copy()
```

> <font color='#074cad'> We start by computing the term count (TC), thus calculating how many times a word appears in each of the race files. </font>

```python
tc = []
races = df.Race.unique()
champions = ['Hylian', 'Zora', 'Goron', 'Gerudo', 'Rito']
N = len(races)


for i in champions:
    with open(f"{tok_path}{i}.txt", 'r') as f: # reading each cleaned race file
        text = f.read()

        words = text.split(' ') # split words

        unique_words = set(words) # get unique words
        
        num_words = dict.fromkeys(unique_words, 0) # create dictionary for unique words
        
        for word in words: # count for unique words
            num_words[word] += 1

        tc.append((num_words)) # save term count in list
                
        print(f'{i}: {sorted(num_words, key=num_words.get, reverse=True)[:10]}') # print 10 most frequent words
```

> <font color='#074cad'> We compute the inverse document frequency (IDF) by finding how many documents each word is found in. </font>

```python
idf = {}

for i in races:
    with open(f"{tok_path}{i}.txt", 'r') as f: # reading each cleaned race file
        text = f.read()

        words = text.split(' ') # split words
        
        for word in words:
            try:
                idf[word].add(i)
            except:
                idf[word] = {i}
for i in idf:
    idf[i] = len(idf[i])

for word, val in idf.items():
        idf[word] = math.log(N / float(val))
```

> <font color='#074cad'> We now have the TC and IDF and define a function to calculate the TC-IDF. </font>

```python
def computeTCIDF(tc, idfs):
    tcidf = {}
    for word, val in tc.items():
        tcidf[word] = val * idfs[word]
    return tcidf
```

```python
# computing the TC-IDF for each race
hylian = computeTCIDF(tc[0], idf)
zora = computeTCIDF(tc[1], idf)
goron = computeTCIDF(tc[2], idf)
gerudo = computeTCIDF(tc[3], idf)
rito = computeTCIDF(tc[4], idf)
```

**Question**
> * Explain the Louvain algorithm and how it finds communities in a newtork.

**Answer** 

<font color='#074cad'>The Louvain algorithm is a greedy optimization algorithm that tries to maximize the modularity $M$. Firstly every node starts in it own community and then the algorithm is divided into to steps.</font>

 


<font color='#074cad'>Step 1 Modularity optimization:</font>

>* <font color='#074cad'> A random order for the nodes is chosen and for each node i, i is moved to the community of a neighbor jthat leads to the biggest positive change in modularity $\Delta M$. If $\Delta M <0$ then i remains in its current comunity. This process is continued until no vertex wants to move.</font>

<font color='#074cad'>Step 2 Community aggregation </font>

>* <font color='#074cad'>The network from step one is being restructered as a weighted network, where each community found in step one is as a single node. The internal edges become loops of weight 2. Meaning the weight on the current internal edges are multiplied with 2. The weights are added to the external edges by summing the number of external edges between the different communities. So the number of edges one community has to another community becomes the weight of the new external edge. </font>

<font color='#074cad'>As the aggregation network has been created the process of step 1 and 2 are repeated. This is done until there canÂ´t be found any improvement in the modularity score. </font>

> <font color='#074cad'> The modularity from the partition is now being calcluated from the undirected graph which consist of the largest connected component </font>

> <font color='#074cad'> The modularity $Q$ is calculated by the following formula: $\sum_{c=1}^{n_c}[\frac{L_c}{L}-(\frac{k_c}{2L})^2]$

> <font color='#074cad'>$L_c$ is the number of links in community c.</font>
    
> <font color='#074cad'>$k_c$ is the total degree of nodes in community c.</font>

> <font color='#074cad'>$L$ is the number of links in the graph.</font>

> <font color='#074cad'> The Modularity score gives an indication of how well the communities partition. The modularity can score between $-1<=M<=1$. If the modularity obtains a positive score it means that the partition has done better on average. A higher score therefore corresponds to more community structure in the network.</font>

```python
L=G.number_of_edges()
M=0
for i in range(len(set(partition.values()))):
    com_list=[]
    for key, value in partition.items():
        df.loc[df.Name==key, "Community"]=value
        if value==i:
            com_list.append(key)
    subg=G.subgraph(com_list).copy()
    Lc=subg.number_of_edges()
    Kc=sum([subg.degree(n) for n in subg.nodes()])
    M+=(Lc/L)-(Kc/(2*L))**2
```

```python
print("Modularity is: ", M)
```

>   * There are many ways to calculate TF-IDF, explain how you've done it and motivate your choices.

**Answer**

> <font color='#074cad'> The TF is calculated by finding the number of times a term or word $t$ appears in a document. This number $t$ is divided by the number of words in the document $d$. Each document or in this case community will therefore have its own term frequency. The term frequency thereby tells us about how often a word occurs in a document.

> <font color='#074cad'> The formula for the TF is as following:</font>

> <font color='#074cad'>$$TF(t,d)=\frac{f_{t,d}}{\sum_k f_{t,d}}$$
</font>

> <font color='#074cad'> When calculating the IDF the logarithm is taken to the number of documents $N$ divided by the number of documents that contain the specific term $n_t$. The IDF therefore only needs to be calculated once as it finds an IDF for each word. </font>

> <font color='#074cad'> The formula for the IDF is:</font>
> <font color='#074cad'>$$IDF(t, D)= log\left(\frac{N}{|\{d\in D:t\in d\}|}\right)$$</font>

> <font color='#074cad'> From the formula it can be seen that if a word doesnÂ´t occur often the IDF score will be large. It can thereby detect rare words across all the documents.</font>

> <font color='#074cad'> When calculatin the TD-IDF it is the TF multiplied by IDF. </font>

> <font color='#074cad'>$$\textit{TF-IDF}(t,d,D) = TF \cdot IDF$$</font>

> <font color='#074cad'>The TF-IDF is used to signify the importance of each word in the document and corpus.</font>

# IDF

> <font color='#074cad'>The IDF is now being calculated for all words in the text consisting of all communities. As mentioned in the description this gives a score based on how rare a word is.
When calculating the IDF The same function as in part 2 is being used. Therefore see part2 for the definition:</font>

```python
def computeIDF(tc, texts):
    N = len(texts)
    
    all_text = set().union(*texts) # list of all unique words
    
    idfDict = dict.fromkeys(all_text, 0) # dict of all unique words
    
    for file in tc: # count how many documents contain the word 
        for word, val in file.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items(): # compute IDF
        idfDict[word] = math.log(N / float(val))
        
    return idfDict
```

```python
idfs=computeIDF(tf, texts)
```

# TF-IDF

```python
def computeTFIDF(tc, idfs):
    tfidf = {}
    for word, val in tc.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

> <font color='#074cad'> In order to align the dialogues with words in LabMT data, the text must be roughly cleaned and tokenized. Next, each character sentiment is calculated as the avg. score of its dialogue. The score is averaged by only including the scores of LabMT recogiced words in order to avoid low sentiment scores in dialogues with only a few recognised words.</font>

```python
sentiment[['Clean_Dialogue','LabMT']]=""

#Calculate avg. dialogue sentiment
for i in range(len(sentiment)):
    tot_sentiment=0
    text=nltk.word_tokenize(sentiment.Dialogue[i].lower())
    text=[lemmatizer.lemmatize(word) for word in text]
    sentiment['Clean_Dialogue'][i]=" ".join(text)
    #Get LabMT scores in dialogue
    rec_words=0.000001
    for ch_word in text:
        if ch_word in list(LabMT.word.values):
            rec_words+=1
            sent_score=LabMT.loc[LabMT.word==ch_word,'happiness_average'].values[0]
            tot_sentiment+=sent_score
    avg_sentiment=tot_sentiment/rec_words
    sentiment['LabMT'][i]=round(avg_sentiment,2)
```
