```python
import os
import networkx as nx
import matplotlib.pyplot as plt
def graphi(char,G):
    pages_path = r"C:/Users/geng8/Desktop/SA/CharactersD/"
    pattern1 = "\[\[(.*?)(?:\|.*?)?\]\]"
    pattern2 ="\{\{(Term|Plural)\|BotW\|(.*?)\}\}" 
    pattern2 ="\{\{Term\|BotW\|(.*?)(?:\|.*?)?\|link|display=name2\}\}"
    count =0
    for node in char:
        
        with open(pages_path+ node + '.txt', ) as file:
            count += 1
            node_description = file.read()
            first = re.findall(pattern1, node_description)
            
            second = re.findall(pattern2, node_description)
            concatenate = list(dict.fromkeys(first + second))
            index1 = np.where(df_characters["Name"] =="{}".format(node))
            G.add_node(node,role ="{}".format(df_characters.role.values[index1]))
            final= [i for i in concatenate if os.path.isfile(pages_path+i +'.txt')== True]
            for j in final:
                index = np.where(df_characters["Name"] =="{}".format(j))
                G.add_node(j,role="{}".format(df_characters.role.values[index]))
                G.add_edge(node, j)
    print(count)
```

```python
import matplotlib.pyplot as plt
from collections import Counter 
node_degree =[]

for i in G.nodes():
    x = G.degree['{}'.format(i)]
    node_degree.append(x)
degree_max,degree_min = np.max(node_degree),np.min(node_degree)
print(degree_max)
print(degree_min)
plt.style.use('seaborn')
num_bin = degree_max - degree_min +1
hist = np.histogram(node_degree,num_bin)
v = np.linspace(degree_min, degree_max, num_bin)#第一个参数为起始点第二个为终止点第三个为等差参数
plt.bar(v, hist[0])#bar 第一个参数为X轴，第二个参数为Y轴.

number= Counter(node_degree)
print(number)

```
